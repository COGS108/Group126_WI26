## Abstract

Player retention and quitting behavior are critical factors in understanding engagement in competitive online games. In this study, we investigate how gameplay outcomes—particularly loss streaks and performance—influence a player’s decision to end a gaming session. Prior research and anecdotal evidence suggest that players are less likely to stop after wins and more likely to disengage following consecutive losses, but the extent to which this behavior can be quantified and predicted remains unclear.

To address this, we analyzed a dataset of gameplay sessions from 984 top Marvel Rivals players, extracting features such as match outcome, performance metrics, skill rating (SR) change, and loss streak length. We first conducted exploratory data analysis and logistic regression to examine statistical relationships between these variables and session termination. We then extended our approach by building several machine learning models—including logistic regression, XGBoost, and random forest classifiers—to evaluate whether quitting behavior could be predicted at the individual match level. Special attention was given to class imbalance, as session-ending events were relatively rare.

Our results show that loss streak length is positively associated with the likelihood of ending a session, supporting the idea that consecutive losses contribute to disengagement. However, across all models, predictive performance remained limited. While models were able to identify some quitting events, they consistently struggled with low precision and recall for session termination, even when using more complex algorithms.

Overall, our results suggest that while loss-streaks and in-game performance contributes to quitting behavior, it is not sufficient to reliably predict when players will stop playing. This indicates that quitting is influenced by additional factors beyond gameplay metrics, such as individual preferences or external constraints.

## Background and Prior Work

Competitive online video games such as Marvel Rivals, a team-based first-person-shooter, provide a natural setting to study how reward and failure influence persistence. Performance statistics are not merely outcomes of play; they engage reinforcement learning mechanisms supported by dopaminergic reward signaling. A common misconception of dopamine is that dopamine neurons always activate upon receiving positive rewards. However, research in Reward Prediction Error (RPE) shows that dopamine neurons don't just respond to any reward: they respond only to rewards that differ from their prediction.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Unexpected negative outcomes, such as losing when one expected to win, can therefore generate strong motivational signals that influence subsequent behavior.

In competitive games, performance feedback is not limited to binary win–loss outcomes. Players also receive more nuanced signals about their performance, including rating changes (such as Skill Rating or SR gains and losses) and individual performance metrics such as kills, deaths, and assists (KDA). These signals provide continuous feedback about whether a player is performing well relative to their expectations. Negative performance signals—such as loss streaks, large rating losses, or poor KDA—may therefore influence whether players continue playing or end their play session.

We focus specifically on high-ranking Marvel Rivals players on the Top 1000 Leaderboard. Because elite players invest substantial time and effort into achieving and maintaining their rank, they are likely to be highly engaged with competitive play and may participate in longer, more sustained gameplay sessions. As a result, their decision to stop playing may be more strongly tied to performance feedback—such as loss streaks, rating losses, or poor in-game performance—rather than external constraints alone. Although this population represents only a small subset of the overall player base, it provides a useful setting for examining how competitive performance signals influence persistence in gameplay.

Prior work suggests that uncertainty and outcome resolution contribute to continued engagement. The transition from uncertainty to resolution can itself produce psychological value, a concept known as "resolution utility."<a name="cite_ref-2"></a><sup>2</sup> <a name="cite_ref-3"></a><sup>3</sup> In gaming contexts, this mechanism may motivate players to continue playing across multiple matches in order to resolve uncertainty about their performance trajectory. However, once outcomes are resolved—particularly when they are repeatedly negative—players may reassess whether continuing to play is worthwhile.

In summary, this project investigates whether recent negative performance signals—such as loss streaks, rating loss magnitude (SR change), and poor KDA—predict the end of a continuous play session among high-ranking Marvel Rivals players.

### Loss Streaks vs Session Behavior

```python
# --- sort properly ---
df = df.sort_values(["anon_player_id", "match_time"]).reset_index(drop=True)

# CONSTRUCT SESSION IDs
SESSION_BREAK_MIN = 45

# first match for each player should always start a new session
new_session = (
    df["idle_gap_min_match_before"].isna() |
    (df["idle_gap_min_match_before"] > SESSION_BREAK_MIN) |
    (df["anon_player_id"] != df["anon_player_id"].shift())
)

# cumulative session number within each player
df["session_number"] = new_session.groupby(df["anon_player_id"]).cumsum()

# optional globally unique session id
df["session_id"] = (
    df["anon_player_id"].astype(str) + "_s" + df["session_number"].astype(str)
)

df["session_end"] = df["session_id"] != df["session_id"].shift(-1)
```

```python
def compute_loss_streak(win_series):
    streaks = []
    current = 0
    for is_win in win_series:
        if not is_win:
            current += 1
        else:
            current = 0
        streaks.append(current)
    return pd.Series(streaks, index=win_series.index)

df["loss_streak"] = (
    df.groupby("session_id")["is_win"]
      .transform(compute_loss_streak)
      .reset_index(level=0, drop=True)
)

df_losses = df[~df["is_win"]].copy()

# collapse rare long streaks into 3+
df_losses["loss_streak_cat"] = df_losses["loss_streak"].apply(
    lambda x: "3+" if x >= 3 else str(int(x))
)

# ordered categories for plotting
order = ["1", "2", "3+"]
df_losses["loss_streak_cat"] = pd.Categorical(
    df_losses["loss_streak_cat"],
    categories=order,
    ordered=True
)
```

```python
plt.figure(figsize=(7, 5))
sns.pointplot(
    data=df_losses,
    x="loss_streak_cat",
    y="session_end",
    order=order,
    errorbar=("ci", 95)
)
plt.xlabel("Consecutive Loss Streak Length")
plt.ylabel("Probability of Session Ending")
plt.title("Effect of Loss Streak Length on Session Termination")
plt.tight_layout()
plt.show()
```

To investigate whether persistent negative outcomes are more strongly associated with session termination than isolated losses, we examined the relationship between consecutive loss streak length and the probability of ending a play session. As shown in the figure above, the likelihood of session termination increased as loss streak length grew. After a single loss, the probability of ending a session was approximately 16.8%. This increased to roughly 19.8% following two consecutive losses and to about 21–22% after three or more losses. While the effect size remains modest, the upward trend suggests that players are more likely to disengage after repeated negative outcomes than after a single unfavorable match.

Additional exploratory analyses examined variables such as match duration and idle time between matches. These factors did not show meaningful differences across session types and were therefore not investigated further.

### Analysis: Effect of Match Outcome on Session Termination

EDA suggested that players are slightly more likely to end their session after losing a match compared to winning. To formally test whether this relationship is statistically significant, we plotted the probability of session ending based on match outcome, then conducted a chi-square test of independence between match outcome and session termination behavior.

### Probability of Session Termination after Win vs Loss

```python
# remove independent matches
df_sessions = df[df["series_categorical"] != "independent match"].copy()

# did the match end the session?
df_sessions["ended_session"] = df_sessions["series_categorical"] == "end series"

# compute probability session ends after win vs loss
result = (
    df_sessions.groupby("is_win")["ended_session"]
    .mean()
    .reset_index()
)

# plot
plt.figure(figsize=(6,5))
sns.barplot(data=result, x="is_win", y="ended_session")

plt.xlabel("Match Outcome")
plt.ylabel("Probability Session Ends")
plt.title("Probability a Session Ends After a Win vs Loss")

plt.xticks([0,1], ["Loss", "Win"])
plt.show()
```

```python
from scipy.stats import chi2_contingency

ct = pd.crosstab(df_sessions["is_win"], df_sessions["ended_session"])

p_end_loss = ct.loc[False, True] / ct.loc[False].sum()
p_end_win = ct.loc[True, True] / ct.loc[True].sum()
diff_prob = p_end_loss - p_end_win

chi2, p, _, _ = chi2_contingency(ct)

summary = pd.DataFrame({
    "Metric": ["After Loss", "After Win", "Difference"],
    "Value": [f"{p_end_loss:.1%}", f"{p_end_win:.1%}", f"{diff_prob:.1%}"]
})

display(summary)

print(f"Chi-square statistic: {chi2:.2f}")
print(f"P-value: {p:.3e}")
```

Sessions ended after 15.9% of losses compared to 13.6% of wins, a 2.3 percentage point increase following losses. A chi-square test confirmed that the relationship between match outcome and session termination is statistically significant (p < 0.001).

Although the difference is statistically significant due to the large number of matches analyzed, the effect size is relatively modest, suggesting that match outcome influences session behavior but is not the sole factor driving session termination.

### Analysis: Logistic Regression with KDA Performance Interaction

To evaluate whether player performance modifies the relationship between match outcome and session termination, we fit a logistic regression model including an interaction term between match outcome and performance score.

```python
import statsmodels.api as sm

df_model = df_sessions.copy()
df_model = df_model.dropna(subset=["performance_score"])
df_model["is_win"] = df_model["is_win"].astype(int)
df_model["interaction"] = df_model["is_win"] * df_model["performance_score"]

X_inter = df_model[["is_win", "performance_score", "interaction"]]
X_inter = sm.add_constant(X_inter)

y = df_model["ended_session"]

interaction_model = sm.Logit(y, X_inter).fit()
print(interaction_model.summary())
```

```python
# make plotting copy
df_inter_plot = df_model.copy()

# predicted probabilities from interaction model
df_inter_plot["pred_prob"] = interaction_model.predict(X_inter)

# aggregate for plotting
inter_plot = (
    df_inter_plot
    .groupby(["performance_level", "is_win"])["pred_prob"]
    .mean()
    .reset_index()
)

inter_plot["result"] = inter_plot["is_win"].map({1: "Win", 0: "Loss"})

# plot
plt.figure(figsize=(6,4))
sns.pointplot(
    data=inter_plot,
    x="performance_level",
    y="pred_prob",
    hue="result",
    errorbar=None
)

plt.ylabel("Predicted Probability Session Ends")
plt.xlabel("Performance Category")
plt.title("Interaction Model: Predicted Session Termination Probability")
plt.tight_layout()
plt.show()
```

A logistic regression model was used to evaluate whether match outcome and player performance predict session termination. The results indicate that winning significantly decreases the probability of ending a session (β = −0.200, p < 0.001). Performance score alone was not a significant predictor (β = −0.008, p = 0.282). 

However, the interaction between match outcome and performance was significant (β = 0.046, p < 0.001), indicating that performance modifies how match outcomes influence session continuation. Consistent with the exploratory analysis, players were most likely to stop playing after losses, particularly when their performance was poor.

### Analysis: Effect of Consecutive Losses on Session Termination

To test whether persistent negative outcomes influence players’ decisions to stop playing, we estimated a logistic regression model predicting whether a match was the final match of a play session. Logistic regression was used because the dependent variable, session termination, is binary.

The model evaluates whether the probability of ending a session increases as players experience longer consecutive loss streaks.

```python
import statsmodels.formula.api as smf

# make numeric binary outcome
df["session_end_num"] = df["session_end"].astype(int)

# optional: restrict to losses only, since this hypothesis is about loss streaks
df_analysis = df[df["is_win"] == False].copy()

model_loss_streak = smf.logit(
    "session_end_num ~ loss_streak",
    data=df_analysis
).fit()

print(model_loss_streak.summary())
```

```python
streak_range = np.arange(1, 6)

predict_df = pd.DataFrame({
    "loss_streak": streak_range
})

predict_df["predicted_prob"] = model_loss_streak.predict(predict_df)


# observed probabilities (reliable streaks only)
observed = (
    df_analysis[df_analysis["loss_streak"] <= 5]
    .groupby("loss_streak")["session_end_num"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(7,5))

# observed
plt.scatter(
    observed["loss_streak"],
    observed["session_end_num"],
    s=80,
    label="Observed Probability"
)

# predicted
plt.plot(
    predict_df["loss_streak"],
    predict_df["predicted_prob"],
    linewidth=2,
    label="Logistic Regression Prediction"
)

plt.xlabel("Loss Streak Length")
plt.ylabel("Probability of Session Ending")
plt.title("Session Termination Probability by Loss Streak Length")

plt.xticks(streak_range)
plt.legend()
plt.grid(alpha=0.3)

plt.show()
```

The regression results indicate that loss streak length was a significant predictor of session termination (β = 0.123, p < 0.001). The positive coefficient suggests that players become increasingly likely to stop playing as consecutive losses accumulate. Converting this coefficient to an odds ratio shows that each additional loss in a streak increases the odds of ending a session by approximately 13%.

The figure above displays both the predicted probabilities from the logistic regression model and the observed probabilities from the data. While the maximum loss streak observed in the dataset was eight matches, fewer than 100 observations occurred for streak lengths greater than five. These rare cases were excluded from the visualization to avoid unstable probability estimates driven by very small sample sizes.

Overall, these results support the hypothesis that session termination is more strongly associated with persistent negative outcomes than with isolated losses.

### Classification Models for Session Termination

Having done all this analysis, can we actually predict when a player is about to quit? To answer this, we built various classification models using features like performance score, loss streak, and match outcomes to classify whether a session ends after each game. This lets us move beyond trends and test whether quitting behavior can be predicted in practice.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import GroupShuffleSplit, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier

#clean NA entries in performance score
df_model = df.dropna(subset=['performance_score']).copy()

#making series categorical a binary representation, end series = 1 (will be class 1 in all ML predictions)
df_model["end_session"] = (df_model["series_categorical"] == "end series").astype(int)

#making temporary X and y for splitting into training test sets
X = df_model.drop(columns=['series_categorical', 'end_session'])
y = df_model['end_session']
groups = df_model['session_id']

#Split indicies into 1 train/test group stratified by session ID
gss = GroupShuffleSplit(n_splits=1, train_size=0.8, random_state=42)
train_idx, test_idx = next(gss.split(X, y, groups))

#make dataframes
train_df = df_model.iloc[train_idx]
test_df = df_model.iloc[test_idx]

#defining predictor features (for normalization later)
numerical_features = ['match_duration', 'SR_change', 'performance_score', 'loss_streak']
categorical_features = ['is_win']

feature_cols = numerical_features + categorical_features

#making X and y for train and test sets
X_train = train_df[feature_cols]
y_train = train_df['end_session']

X_test = test_df[feature_cols]
y_test = test_df['end_session']
```

```python
#function for plotting confusion matrix
def plot_confusion_matrix(y_true, y_pred, title="Confusion Matrix"):
    cm = confusion_matrix(y_true, y_pred)
    
    # Convert to percentages
    cm_percent = cm / cm.sum(axis=1, keepdims=True)
    
    # Create labels with count + %
    labels = np.array([
        [f"{cm[i,j]}\n({cm_percent[i,j]*100:.1f}%)" for j in range(cm.shape[1])]
        for i in range(cm.shape[0])
    ])
    
    plt.figure(figsize=(6,5))
    
    sns.heatmap(
        cm,
        annot=labels,
        fmt="",
        cmap="crest",   # nicer than default blue
        cbar=False,
        linewidths=1
    )
    
    plt.xlabel("Predicted", fontsize=12)
    plt.ylabel("Actual", fontsize=12)
    plt.title(title, fontsize=14)
    
    plt.xticks([0.5, 1.5], ["Continue", "End"])
    plt.yticks([0.5, 1.5], ["Continue", "End"], rotation=0)
    
    plt.tight_layout()
    plt.show()
```

```python
#logistic regression, fairly simple model that assumes linear relationship btw feature and outcome
#preprocessing to normalize numeric and categorical vars
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ]
)

#initalize pipeline for gridsearch, Logistic regression set to 1000 to get a better fit, weighting none for default
pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('clf', LogisticRegression(max_iter=1000, class_weight=None))
])

#hyperparameters to tune, loose to strict C and different weightings to compensate for unabalanced dataset
param_grid = {
    'clf__C': [0.1, 1, 10],
    'clf__class_weight': ['balanced', {0: 1, 1: 2}, {0: 1, 1: 6}]
}

#grid search across hyperparameters
#parameters are same across all future grid searches, 5 fold cross validation, f1_macro scoring to find optimal model 
#(because far fewer end series samples so accuracy will just return a model that classifies everything in-series)
grid = GridSearchCV(pipe, param_grid, cv=5, scoring='f1_macro')
grid.fit(X_train, y_train)

#grab best model
best_model = grid.best_estimator_

y_pred = best_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
plot_confusion_matrix(y_test, y_pred, title="Logistic Regression")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

Using a logistic regression model optimized for balanced performance, we were able to identify approximately 42% of session-ending events. However, the model also produced many false positives, indicating that while quitting behavior is partially predictable, it is influenced by factors beyond the gameplay features included in our model.

```python
#identical process to first, but different model and different hyperparameters
#XGBClassifer makes a large number of trees sequentially, each aiming to correct the last, and returns the weighted sum
#more complex model that doesn't assume linear relationship
!pip install xgboost
from xgboost import XGBClassifier

pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('clf', XGBClassifier(eval_metric='logloss'))
])

#Hyperparameters in order are:
#variable numbers of trees to train (more trees = more fit)
#different depths of each tree (more depth = more fit)
#different contribution values of each tree (how much each tree matters, higher = more fit)
#different balancings of positive and negative weights (weighting for how much model cares about incorrectly predicting class 1 (our end series))
param_grid = {
    'clf__n_estimators': [100, 200],
    'clf__max_depth': [3, 6],
    'clf__learning_rate': [0.01, 0.1],
    'clf__scale_pos_weight': [1, 6]  # 6 helps with your 13.8% minority class
}
grid = GridSearchCV(pipe, param_grid, cv=5, scoring='f1_macro')
grid.fit(X_train, y_train)

best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
plot_confusion_matrix(y_test, y_pred, title="XGBoost (Extreme Gradient Boosting)")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

We also tested a more advanced model, XGBoost, to see if nonlinear relationships would improve prediction. While it slightly increased overall accuracy, it did not meaningfully improve our ability to detect quitting behavior. This suggests that the limitation lies not in the model, but in the available features.

```python
#random forest classifier
#similar to XBG but makes all trees at the same time independently, less sensitive to noise
pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('clf', RandomForestClassifier(random_state=42))
])

#Hyperparameters are:
#number of parallel trees
#levels of each tree
#weighting of categories (balanced makes the model try to weight automatically according to sample size ratios)
param_grid = {
    'clf__n_estimators': [100, 200],
    'clf__max_depth': [None, 10, 20],
    'clf__class_weight': ['balanced', 'balanced_subsample']
}


grid = GridSearchCV(pipe, param_grid, cv=5, scoring='f1_macro', n_jobs=-1)
grid.fit(X_train, y_train)

best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)

print(f"Best Params: {grid.best_params_}")
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
plot_confusion_matrix(y_test, y_pred, title="Random Forest Classifier")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

While the Random Forest model achieved the highest accuracy (68%), this improvement is largely driven by predicting continued play, the majority class. Its relatively low performance in detecting session-ending events suggests that quitting behavior is still difficult to predict from gameplay features alone.

Across all models, predicting when players quit remains challenging, suggesting that quitting behavior is influenced by factors beyond the gameplay features included in our analysis.

## Discussion and Conclusion

This study examined how negative performance signals—particularly loss streaks and in-game performance—predict session termination among high-ranking Marvel Rivals players. Across our analyses, loss streak length emerged as the most consistent and robust predictor of session endings, with probability increasing from approximately 13.6% at a loss streak of 0 to 24% at longer streaks (e.g., 4–5 losses). Logistic regression results confirmed this relationship as statistically significant, and visualization of predicted probabilities showed a clear upward trend. In contrast, performance score alone had a relatively weak direct effect, though its influence became more meaningful when considered alongside win–loss outcomes, suggesting that performance matters more when it reinforces negative results.

These findings closely align with the theoretical framework outlined in our background. Reward Prediction Error (RPE) theory suggests that dopamine responses are strongest when outcomes deviate from expectations, particularly in the negative direction. In this context, loss streaks likely amplify negative prediction errors, as repeated unexpected losses signal to the player that their expectations about performance or matchmaking are being violated. This may explain why isolated losses are often tolerated, while consecutive losses produce a disproportionately strong behavioral response—namely, quitting. Our results therefore extend prior neuroscience findings into a real-world behavioral setting, showing how cumulative negative prediction errors can shape persistence decisions in competitive environments.

At the same time, our results can also be interpreted through the lens of resolution utility. Players may initially continue playing through uncertainty—attempting to “resolve” whether a loss was a fluke or part of a downward trend. However, once that uncertainty is resolved through repeated negative outcomes, the psychological value of continuing diminishes. In other words, loss streaks may act as a tipping point, where players transition from motivated engagement to disengagement. This helps explain why players do not immediately quit after a single loss but are increasingly likely to stop after several in a row, reflecting a shift from uncertainty-driven persistence to outcome-driven withdrawal.

Taken together, our findings suggest that quitting behavior is not simply a failure of persistence, but rather a structured and adaptive response to negative feedback loops. Even among highly competitive Top 1000 players—who are presumably more motivated and resilient—session termination appears to follow predictable behavioral patterns. This challenges the common belief that “serious” players will always push through losses; instead, even elite players appear sensitive to accumulated negative signals and adjust their behavior accordingly.

However, some results were less intuitive. The interaction between performance and game outcome revealed a nuanced pattern: low performance paired with a win was associated with the lowest probability of ending a session (12.6%), whereas low performance paired with a loss produced the highest probability of quitting (15.7%). At first glance, this seems contradictory: if low performance contributes to frustration, one might expect it to consistently increase quitting behavior. Instead, its effect appears to depend heavily on context. One possible explanation is that when players perform poorly but still win, they may feel motivated to continue playing in order to “redeem” themselves and prove their individual competence in subsequent matches. Alternatively, players may derive enough satisfaction from the win itself—even if they were not the primary contributor—to offset frustration, suggesting that being “carried” can still reinforce continued engagement. This highlights that performance signals are not interpreted in isolation, but rather through the lens of outcome, making outcome-based signals like loss streaks more consistently influential predictors of quitting behavior.

Several limitations should be considered. First, our dataset focuses exclusively on Top 1000 players, which may limit generalizability to the broader player base. These players likely have different motivations, time investments, and tolerance for losses compared to casual players. Second, our analysis is observational and does not establish causality; while loss streaks are strongly associated with quitting, other unobserved factors—such as fatigue, time constraints, or emotional state—may also contribute. Third, certain variables, including “unknown” role classifications and aggregated performance metrics, may reduce measurement precision. Performance scores across different roles may not be directly comparable, introducing noise into the analysis. These factors may explain why performance alone was not a strong predictor, while outcome-based signals like loss streaks were much more influential. Finally, the dataset lacks contextual information such as whether players were solo queueing or playing with friends, which could meaningfully affect persistence.

Despite these limitations, our findings have important practical implications. For game developers, understanding that loss streaks significantly drive disengagement suggests opportunities to improve player retention. For example, matchmaking systems could be adjusted to reduce prolonged loss streaks, or games could incorporate interventions such as performance-based feedback, encouragement systems, or subtle difficulty adjustments to interrupt negative feedback loops. On the other hand, for players, these findings highlight a form of behavioral self-awareness: recognizing that the urge to “keep playing until you win” may actually be driven by accumulating negative signals rather than rational decision-making. Understanding this pattern may help players avoid emotional traps, manage tilt, and make more deliberate choices about when to stop.

Future work could extend this analysis in several directions. One particularly valuable addition would be to examine session length directly, by calculating average session duration and testing whether the presence of loss streaks leads to significantly shorter sessions. This would provide an additional behavioral measure supporting the importance of loss streaks. Longitudinal data could also help capture how player behavior evolves over time, while more granular performance metrics tailored to specific roles could improve measurement accuracy. Incorporating contextual variables—such as time of day, social play, or prior session outcomes—would further enrich the analysis and help disentangle competing explanations.

In conclusion, this study demonstrates that session termination in competitive gaming is not random, but systematically influenced by patterns of negative outcomes. By linking loss streaks to quitting behavior and grounding these findings in theories of reward prediction error and resolution utility, we provide a more comprehensive understanding of how players respond to success and failure. Ultimately, these results suggest that quitting is not merely “giving up,” but rather a predictable and often rational response to sustained negative feedback—one that reflects broader principles of human decision-making under uncertainty.
