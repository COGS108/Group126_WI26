```python
# converting the json output of the request to a dictionary of dictionaries
tempJsonHolder = json.loads( r.text )['match_history']
```

```python
# initializing list to hold several lists of dictionaries with data from each match
allMatchList = []
for x in tempJsonHolder:
    # init list to hold single match data, grabbing play duration, time stamp match was played, rr change, and a boolean for win/loss (T = win, F = loss)
    singleMatchList = []
    # assigning indexed match to variable match
    match = x
    singleMatchList.append({
        # mpd and mts are directly indexable
        "match_play_duration": match.get("match_play_duration"),
        "match_time_stamp": match.get("match_time_stamp"),
        # AS and IW are in the match player dictionary, AS in score_info and IW in is_win
        "add_score": match.get("match_player", {}).get("score_info", {}).get("add_score"),
        "is_win": match.get("match_player", {}).get("is_win", {}).get("is_win")
    })
    allMatchList.append( singleMatchList )
```
