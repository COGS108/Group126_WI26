```python
# some basic formatting of the above step could be
import requests
import json
import time

API_KEY = "d1a23ed65bdf9edb2ea6697ae30976bc5c1722228f91d63298b25bfd8b6043f6"

with open("./data/marvel_rivals_names.txt", "r") as f:
    list_of_uids = f.read().splitlines()

all_raw_data = {}

for i, player in enumerate(list_of_uids, start=1):
    endpoint = f"https://marvelrivalsapi.com/api/v2/player/{player}/match-history?season=6&limit=100&game_mode=2"
    
    try:
        response = requests.get(
            endpoint,
            headers={'x-api-key': API_KEY},
            timeout=15
        )
        
        if response.status_code == 200:
            all_raw_data[player] = response.json()
            print(f"[{i}/{len(list_of_uids)}] Success: {player}")
        else:
            print(f"[{i}/{len(list_of_uids)}] Failed ({response.status_code}): {player}")
    
    except Exception as e:
        print(f"[{i}/{len(list_of_uids)}] Error for {player}: {e}")
    
    time.sleep(0.2)  

with open("./data/marvel_rivals_raw.json", "w") as f:
    json.dump(all_raw_data, f, indent=4)

print("All raw JSON saved successfully.")

# finally run every item in the Json list through one of the above functions and add it to a new list of the simplified data
# you will end up with a list that holds lists that holds lists of dictionaries with singular keys
# or a list that holds lists of single dictionaries with several keys
# depending on method
#
#
# feel free to convert the simplifiers to actual functions that can be applied to all elements of a list or simplify/optimize them
```
